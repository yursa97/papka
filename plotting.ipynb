```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('I-two-tone_raw_data.pkl', 'rb') as fhandle:
    data = pickle.load(fhandle)
```

```python
data
```

```python
voltage = data['AWG voltage [V]']
transm = data['data']
freq = data['frequency']
```

```python
plt.pcolormesh(voltage, freq/1e9, np.abs(transm.T)*1e3,cmap='RdBu_r')
plt.xlabel("AWG voltage [V]")
plt.ylabel("Frequency [GHz]")
z = plt.colorbar()
plt.grid(color='black',linestyle='--', linewidth=0.5)
z.set_label('Transmission amplitude [a.u.]', rotation=90)
plt.savefig('STS.pdf')
```
